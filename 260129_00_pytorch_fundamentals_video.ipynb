### Random tensors (21)

Why random tensors?

Random tensors are important because the way many nueral networks learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data.

`Start with random numbers -> look at data -> update random numbers -> look at data`

- 왜 랜덤텐서를 사용하는가   
  - 랜덤텐서가 학습 초기값인데 이거를 조금씩 다르게 줘야 학습이 다양하게 됨.   
    예를 들어서 0으로 똑같으면 다 같은 결과값을 냄

```python
# Create a random tensor of size (3, 4)
random_tensor = torch.rand(3, 4)
random_tensor
```

```python
random_tensor.ndim
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(224, 224, 3))
random_image_size_tensor
```
